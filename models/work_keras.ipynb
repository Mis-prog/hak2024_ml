```python
from keras.preprocessing import image
from keras.applications.imagenet_utils import decode_predictions, preprocess_input
from keras.models import Model
import keras
import cv2
import os
import shutil
from PIL import Image
import numpy as np
import pandas as pd
```

```python
def load_image(path):
    img = image.load_img(path, target_size=model.input_shape[1:3])
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    return img, x
```

```python
model = keras.applications.vgg16.VGG16(weights='imagenet', include_top=True)
```

```python
feat_extractor = Model(inputs=model.input, outputs=model.get_layer("fc2").output)
```

```python
def get_vector(root_folder):
    vectors = {}
    for root, dirs, files in os.walk(root_folder):
        current_feature = []
        dir_name="" 
        for file in files:
            if file.endswith(".jpg"):
                image_path = os.path.join(root, file)
                img, x = load_image(image_path)
                feat = feat_extractor.predict(x)[0]
                dir_name = os.path.basename(root)
                current_feature.append(feat)

        vectors[dir_name] = np.array(np.mean(current_feature,axis=0))
        # print(dir_name,'\n')
    # data = pd.DataFrame.from_dict(vectors, orient='index')
    # data.to_csv("blogger_vector.csv", index=True)
    return vectors
```

```python
vectors=get_vector("bloggers_cropped_new")
```

```python
del vectors[""]
```

```python
data=pd.DataFrame(vectors)
data.to_csv("bloggers_vector.csv",index="False")
```
